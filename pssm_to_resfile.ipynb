```python
def pssm_2_resfile(pssmfile, remove_res_list, 
                   probability_threshold, uncertainty_threshold = 0.8, set_structure_start=0): 
    #(infile, list, float) --> outfile (Rosetta resfile format)
    """
    read in a PSSM file (transfac format), a list of residues that are in the sequence alignment but not in the structure
    #(i.e. gaps that you've removed), and a probability threshold (this is a float; only probabilities above this threshold
    #will be output into the resfile; if no probabilites above threshold, NATAA will be written (default to 
    the maximum likelihood residue) and a comment made). 
    """
    
    
    #here we read the data in
    data = open(pssmfile, 'r')
    lines = data.readlines() 
    list_by_aa_site = [i.split('\t') for i in lines]
    
    #we clean the data so its a nice list of lists
    cleaned_list_by_aa_site = []
    for site_list in list_by_aa_site:
        clean_site_list = []
        for probability in site_list :
            clean_Pr = probability.strip()
            clean_site_list.append(clean_Pr)
        cleaned_list_by_aa_site.append(clean_site_list)

    #and here we start making our resfile
    out_resfile = open(str(pssmfile[:-9])+'.resfile', 'w+')
    
    out_resfile.write('NATAA \n')
    out_resfile.write('USE_INPUT_SC \n')
    out_resfile.write('start \n')
    
    aa = {1:'A', 2:'R', 3:'N', 4:'D', 5:'C', 6:'Q', 7:'E', 8:'G', 9:'H',10:'I',
               11:'L', 12:'K', 13: 'M', 14:'F', 15:'P', 16:'S', 17:'T',
               18:'W', 19:'Y', 20:'V'}
    
    chosen_cleaned_list_by_aa_site= []
        
    uncertainty_check = 0
    for i in range(len(cleaned_list_by_aa_site)):
        if i not in remove_res_list:
            chosen_cleaned_list_by_aa_site.append(cleaned_list_by_aa_site[i])
            #tested, above code does its job
    
   # print chosen_cleaned_list_by_aa_site
    
    #outsequence is just used to check if the program is working correctly (and that inputs yield correct results). If
    #the desired sequence is recapitulated, then the program and inputs are working as desired. 

    
    out_sequence = []
    for i in chosen_cleaned_list_by_aa_site[1:]:
        aminos = []
        aa_indexes = []
        res_id = chosen_cleaned_list_by_aa_site.index(i) + set_structure_start
        probabilities = []
        
       
        #print chosen_cleaned_list_by_aa_site
        check_list = [float(k) for k in i[1:] if float(k)>=uncertainty_threshold]
        print check_list
        if len(check_list) == 0:
            uncertainty_check = uncertainty_check + 1
            

        
        
        
        for j in range(len(i)): 
            #print float(i[j])
            

            
            #conditions: probability cannot be 0.0; probability must be of type float; probability must be 
            #greater than or equal to probability threshold. 
            
            if (i[j] != '0.0') and ('.' in i[j]) and (float(i[j]) >= probability_threshold):
                aa_indexes.append(j)
            probabilities.append(float(i[j]))
        #print probabilities
        

        
                  
                  
        max_pr_res = aa.get(probabilities.index(max(probabilities[1:])))

        for index in aa_indexes: 
            aminos.append(aa.get(index))
        #print aa_indexes
        #print aminos

        aa_space = "".join(aminos)
        out_sequence.append(max_pr_res)
        
        
        out_resfile.write(str(res_id) + " A " + " PIKAA " +  str(aa_space) + ' #'+str(max_pr_res) +  "\n")
        
    
    out_sequence_str = "".join(out_sequence)
    #print out_sequence_str
    
    out_resfile.write('# ' + out_sequence_str)
                
                
                
    #function tested, works perfectly!            
                
                
        
    print uncertainty_check
        
        
        
    
    
    


    
```
